# Class & Object

```python
%%writefile class.cpp


#include <iostream>
using namespace std;

class Car {
public:
    string brand;
    int year;

    void showInfo() {
        cout << "Brand: " << brand << ", Year: " << year << endl;
    }
};

int main() {
    Car car1;
    car1.brand = "Toyota";
    car1.year = 2020;

    Car car2;
    car2.brand = "Honda";
    car2.year = 2022;

    car1.showInfo();
    car2.showInfo();

    return 0;
}
```

```python
!g++ class.cpp -o class
```

```python
!./class
```

# Encapsulation

```python
%%writefile encapsulation.cpp
#include <iostream>
using namespace std;

class Car {
private:
    string brand;
    int year;

public:
    void setData(string b, int y) {
        brand = b;
        if (y > 1885)
            year = y;
        else
            year = 0;
    }

    void showInfo() {
        cout << "Brand: " << brand << ", Year: " << year << endl;
    }
};

int main() {
    Car car1, car2;

    car1.setData("Toyota", 2020);
    car2.setData("Honda", 2022);

    car1.showInfo();
    car2.showInfo();

    return 0;
}
```

```python
!g++ encapsulation.cpp -o encapsulation
```

```python
!./encapsulation
```

# Overloading

```python
%%writefile overloading.cpp
#include <iostream>
using namespace std;

class Math {
public:
    int add(int a, int b) {
        return a + b;
    }

    int add(int a, int b, int c) {
        return a + b + c;
    }

    double add(double a, double b) {
        return a + b;
    }
};

int main() {
    Math m;
    cout << "Sum of 2 ints: " << m.add(5, 10) << endl;
    cout << "Sum of 3 ints: " << m.add(5, 10, 15) << endl;
    cout << "Sum of 2 doubles: " << m.add(3.5, 2.2) << endl;
    return 0;
}
```

```python
!g++ overloading.cpp -o overloading
```

```python
!./overloading
```

# Inheritence

```python
%%writefile inheritance.cpp
#include <iostream>
using namespace std;

class Person {
public:
    string name;
    int age;

    void introduce() {
        cout << "I am " << name << ", Age: " << age << endl;
    }
};

class Student : public Person {
public:
    string school;

    void study() {
        cout << name << " is studying at " << school << endl;
    }
};

int main() {
    Student s1;
    s1.name = "Aarav";
    s1.age = 20;
    s1.school = "Delhi University";

    s1.introduce();
    s1.study();

    return 0;
}
```

```python
!g++ inheritance.cpp -o inheritance
```

```python
!./inheritance
```

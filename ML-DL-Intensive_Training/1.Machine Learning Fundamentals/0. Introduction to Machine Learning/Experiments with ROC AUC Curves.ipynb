```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
d=pd.DataFrame({'score':np.random.random(size=100)})
d['target']=(d['score']>0.3).astype(int)
```

```python
sns.lineplot(x='score',y='target',data=d)
```

```python
TPR=[]
FPR=[]
real=d['target']
for cutoff in np.linspace(0,1,100):
    predicted=(d['score']>cutoff).astype(int)
    TP=((real==1)&(predicted==1)).sum()
    FP=((real==0)&(predicted==1)).sum()
    TN=((real==0)&(predicted==0)).sum()
    FN=((real==1)&(predicted==0)).sum()
    
    TPR.append(TP/(TP+FN))
    FPR.append(FP/(TN+FP))

temp=pd.DataFrame({'TPR':TPR,'FPR':FPR})

sns.lmplot(y='TPR',x='FPR',data=temp,fit_reg=False)
```

```python
##Typecasting 
ld_train['Amount.Requested'] = ld_train['Amount.Requested'].fillna(0, inplace=True)
ld_train['Amount.Requested'] = ld_train['Amount.Requested'].astype('float')
```

```python
def convert_percent(val):
    val = str(val)
    new_val = val.replace('%', '')
    return float(new_val)


##

ld_train['Debt.To.Income.Ratio'] = ld_train['Debt.To.Income.Ratio'].apply(convert_percent)
ld_train
```

```python
import numpy as np
```

```python
def convert_emp_length(val):
    if type(val) == str:
        if val == '10+ years':
            return 10
        elif val == '< 1 year':
            return 0
        else:
            return int(val.split()[0])
    else:
        return None
        
ld_train['Employment.Length'] = ld_train['Employment.Length'].apply(convert_emp_length)
```

```python
mean_value = ld_train['Employment.Length'].mean()

ld_train['Employment.Length'].fillna(mean_value, inplace=True)
```

```python
# mean_imputation
df["Age"] = df["Age"].fillna(df["Age"].mean())
df
```

```python
# median_imputation
df["Salary"] = df["Salary"].fillna(df["Salary"].median(), inplace=True)
df
```

```python
# mode_imputation
df["Experience (Years)"] = df["Experience (Years)"].fillna(df["Experience (Years)"].mode(),inplace=True)
df
```

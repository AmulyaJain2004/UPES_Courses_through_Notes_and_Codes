```python
df_country = pd.read_excel('Country-Code.xlsx')
df_country.head()
```

```python
df.columns
```

```python
final_df = pd.merge(df, df_country, on='Country Code', how='left')
```

```python
final_df.head(2)
```

```python
final_df.dtypes
```

```python
final_df.columns
```

```python
# Counting the records according to the country-wise
final_df.Country.value_counts()
```

### This shows Zomato is mostly available in India.

```python
# Names of countries
country_names=final_df.Country.value_counts().index
```

```python
final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
country_val
```

```python
# Pie Chart- Top 5 Countries that uses Zomato
plt.pie(country_val[:5],labels=country_names[:5], autopct='%1.2f%%') 
```

Observation:
- Zomato maximum transaction or records are from India
- After that USA and then United Kingdom. 

```python
final_df.columns
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size()
```

```python
final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index()
```

```python
ratings = final_df.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index(name='Ratings Count')
```

```python
ratings
```

## Observations:
1. When rating is between 4.5 to 4.9 --> Excellent
2. When rating is between 4.0 to 3.4 --> Very Good
3. When rating is between 3.5 to 3.9 --> Good
4. When rating is between 3.0 to 3.4 --> Average
5. When rating is between 2.5 to 2.9 --> Average 
6. When rating is between 2.0 to 2.4 --> Poor

```python
ratings.head()
```

```python
matplotlib.rcParams['figure.figsize'] = (15, 9)
sns.barplot(x="Aggregate rating", y="Ratings Count", data=ratings, hue='Rating color', palette=['blue', 'red', 'orange', 'yellow', 'green','green'])
```

Observations:
1. Not Rated count is very high
2. Maximum number of ratings are between 2.5 to 3.4

```python
# Count plot
sns.countplot(x="Rating color", hue='Rating color', data=ratings, palette=['blue', 'red', 'orange', 'yellow', 'green','green'])
```

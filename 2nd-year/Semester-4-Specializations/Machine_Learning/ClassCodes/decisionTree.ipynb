```python
from sklearn.datasets import load_iris
from sklearn import tree
iris = load_iris()
X, y = iris.data, iris.target
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
tree.plot_tree(clf)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(X, columns=iris.feature_names)
df['target'] = y
```

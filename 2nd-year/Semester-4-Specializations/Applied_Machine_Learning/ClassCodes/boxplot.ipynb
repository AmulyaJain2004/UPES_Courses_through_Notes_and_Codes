```python
import numpy as np
import pandas as pd
import matplotlib . pyplot as plt
import seaborn as sns
from scipy . stats import skew
```

```python
# Generate a synthetic dataset
np . random . seed (42)
data = np . concatenate ([ np . random . normal (50 , 10 , 100) , np . random .
normal (100 , 20 , 50) , np . random . normal (200 , 50 , 10) ])
df = pd . DataFrame ( data , columns =[ 'Values'])
# Display the first few rows of the dataset
print ( df . head () )
```

```python
# Calculate skewness
skewness = skew ( df ['Values'])
print (f" Skewness of the data : { skewness :.4f}")
```

```python
# Plot the histogram
plt . figure ( figsize =(10 , 6))
sns . histplot ( data = df ['Values'], kde = True, color='blue')
plt.title('Data Distribution with Skewness')
plt . xlabel ('Values ')
plt . ylabel ('Frequency')
plt.show()
```
